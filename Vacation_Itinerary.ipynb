```python
# Import the dependencies.
import pandas as pd
import gmaps

# Import the API key.
from config_gmaps import g_key

```

```python


# Read the file and store it in a Pandas DataFrame.
vacation_data_df = pd.read_csv('weather_data/WeatherPy_vacation.csv')
vacation_data_df.head(100)
```

```python
India_itinerary_df = vacation_data_df.loc[(vacation_data_df['Country'] =='IN')]
India_itinerary_df
```

```python
# choose four cities from Brazil.(Cidreira, Santa cruz do sul, Cangusu, Rio grande)
four_cities_df = India_itinerary_df.loc[[30,43,95,178],:]
four_cities_df
```

```python

# create a list of tuple of 4 cities' coordinate.
gmaps.configure(api_key = g_key)

India_coord_list = list(zip(four_cities_df['Lat'], four_cities_df['Lng']))
India_coord_list
```

```python
# create a directions layer map by gmaps directions layer.
fig = gmaps.figure()
India_route = gmaps.directions_layer(India_coord_list[0],India_coord_list[-1], waypoints=[India_coord_list[1],India_coord_list[2]],
                                  travel_mode='BICYCLING')
fig.add_layer(India_route)

fig
```

```python
# create India FOUR cities marker layer map with a pop-up info box 
gmaps.configure(api_key = g_key)
locations = four_cities_df[['Lat','Lng']]
info_box_template='''
<dl>
<dt>Hotel Name</dt><dd>{Hotel Name}</dd>
<dt>City</dt><dd>{City}</dd>
<dt>Country</dt><dd>{Country}</dd>
<dt>Current Weather</dt><dd>{Current Description} and {Max Temp}°F</dd>
</dl>'''

hotel_info = [info_box_template.format(**row) for index, row in four_cities_df.iterrows() ]
fig = gmaps.figure()
marker_layer = gmaps.marker_layer(locations, info_box_content = hotel_info)
fig.add_layer(marker_layer)

fig
```
